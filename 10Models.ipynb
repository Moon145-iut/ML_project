## Preprocessing: Normalizing text, Deduplicate groups, and Handling ambiguous duplicates



```python
import unicodedata, re
from collections import defaultdict

# PARAMETERS - CHANGE if you want
AMBIGUOUS_STRATEGY = 'drop_ambiguous'  # options: 'drop_ambiguous', 'keep_first', 'inspect'
LOWERCASE = True # whether to lowercase text during dedup grouping and ML pipeline
PRINT_AMBIG = True

def normalize_text(t):
    if not isinstance(t, str):
        t = str(t)
    # unicode normalize
    t = unicodedata.normalize('NFKC', t)
    # collapse whitespace
    t = re.sub(r'\s+', ' ', t).strip()
    if LOWERCASE:
        t = t.lower()
    return t

# Add normalized text column (used for duplicate grouping)
df['text_norm'] = df['text'].apply(normalize_text)

# Group identical normalized texts
groups = df.groupby('text_norm').agg({'label': lambda s: list(s), 'text': lambda s: list(s)}).reset_index()
groups['n_samples'] = groups['label'].apply(len)
groups['unique_labels'] = groups['label'].apply(lambda lab_list: sorted(set(lab_list)))
print('Total unique text groups:', len(groups))
# find ambiguous groups
ambig = groups[groups['unique_labels'].apply(lambda x: len(x) > 1)]
print('Ambiguous duplicate groups (same text, multiple labels):', len(ambig))
if PRINT_AMBIG and len(ambig)>0:
    display(ambig.head(10))

if len(ambig)>0:
    if AMBIGUOUS_STRATEGY == 'inspect':
        raise SystemExit("Ambiguous duplicate groups detected. Set AMBIGUOUS_STRATEGY to drop_ambiguous or keep_first after manual inspection.")
    elif AMBIGUOUS_STRATEGY == 'drop_ambiguous':
        # drop all rows whose normalized text is in ambiguous set
        ambig_texts = set(ambig['text_norm'])
        before = len(df)
        df = df[~df['text_norm'].isin(ambig_texts)].reset_index(drop=True)
        print(f"Dropped {before - len(df)} rows from ambiguous duplicate groups.")
    elif AMBIGUOUS_STRATEGY == 'keep_first':
        # keep only the first occurrence per text_norm
        df = df.drop_duplicates(subset=['text_norm'], keep='first').reset_index(drop=True)
        print("Kept first occurrence of each duplicated normalized text (including ambiguous).")
    else:
        raise ValueError("Unknown AMBIGUOUS_STRATEGY")

# Now create group ids for each unique normalized text (non-ambiguous groups)
unique_texts = sorted(df['text_norm'].unique())
text_to_group = {t:i for i,t in enumerate(unique_texts)}
df['group_id'] = df['text_norm'].map(text_to_group)
print('After dedup handling, rows:', len(df), 'unique groups:', len(unique_texts))
```

## Train-Test Splitting
#### **Strategy**:
###### Stratified K-fold Cross Validation but **keeping 15% dedicated for Testing**
#####**We implemented Held-Out Test 15% rule. Then the remaining 85% will be used with 5-fold CV for model selection**

```python
import numpy as np
from sklearn.model_selection import GroupShuffleSplit

# Define a writable working directory
WORKDIR = "/content/drive/MyDrive/symptom2disease_project"

# create if not exists
os.makedirs(WORKDIR, exist_ok=True)

TEST_GROUP_FRAC = 0.15  # fraction of total groups to hold out
random_state = 42

# Build group-level label: each group should have a single label now
group_df = df.groupby('group_id').agg({'label': lambda s: s.iloc[0], 'text': lambda s: s.iloc[0]}).reset_index()
group_df['label_id'] = group_df['label'].map({lab:i for i,lab in enumerate(sorted(df['label'].unique()))})

# StratifiedGroupKFold for stratified group split
try:
    from sklearn.model_selection import StratifiedGroupKFold, StratifiedShuffleSplit
    # StratifiedGroupKFold available: use StratifiedShuffleSplit-like behavior for holdout
    from sklearn.model_selection import train_test_split
    # We'll select test groups via stratified sampling across group_df labels
    _, test_idxs = train_test_split(group_df.index, test_size=TEST_GROUP_FRAC, stratify=group_df['label_id'], random_state=random_state)
    test_group_ids = set(group_df.loc[test_idxs, 'group_id'].tolist())
    print('Used stratified sampling for holdout (via train_test_split).')
except Exception as e:
    # Fallback: To use GroupShuffleSplit (random) BUT ensure label distribution roughly preserved
    print('StratifiedGroupKFold not available; using greedy GroupShuffleSplit fallback.')
    from sklearn.model_selection import GroupShuffleSplit
    gss = GroupShuffleSplit(n_splits=1, test_size=TEST_GROUP_FRAC, random_state=random_state)
    train_idx, test_idx = next(gss.split(group_df, groups=group_df['group_id']))
    test_group_ids = set(group_df.loc[test_idx, 'group_id'].tolist())

# Building final trainval and final_test sets (keep all rows with groups)
final_test_df = df[df['group_id'].isin(test_group_ids)].reset_index(drop=True)
trainval_df = df[~df['group_id'].isin(test_group_ids)].reset_index(drop=True)
print('Final test rows:', len(final_test_df), 'Train+Val rows:', len(trainval_df))

# Saving to WORKDIR for reproducibility
final_test_df.to_csv(os.path.join(WORKDIR, 'final_test.csv'), index=False)
trainval_df.to_csv(os.path.join(WORKDIR, 'trainval.csv'), index=False)
print('Saved trainval.csv and final_test.csv in WORKDIR')
```

## K-fold Cross Validation

#### **StratifiedGroupKFold, otherwise GroupKFold**

```python
import numpy as np
import random
import torch

# Set seed for reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
```

```python
from sklearn.model_selection import GroupKFold, StratifiedKFold

def get_stratified_group_kfold_splits(df, n_splits=5, random_state=SEED):
    # df must contain 'group_id' and 'label'
    groups = df['group_id'].values
    labels = df['label'].map({lab:i for i,lab in enumerate(sorted(df['label'].unique()))}).values
    unique_groups = np.unique(groups)
    # Make group-level arrays
    group_to_label = {}
    for g in unique_groups:
        # pick label of first sample in group
        lab = df[df['group_id']==g]['label'].iloc[0]
        group_to_label[g] = lab
    group_labels = np.array([group_to_label[g] for g in unique_groups])
    group_label_ids = np.array([list(sorted(df['label'].unique())).index(x) for x in group_labels])
    try:
        # scikit-learn >=1.1 has StratifiedGroupKFold
        from sklearn.model_selection import StratifiedGroupKFold
        sgkf = StratifiedGroupKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
        # sgkf splits group indices: need to map groups back to row indices
        for train_group_idx, val_group_idx in sgkf.split(unique_groups, group_label_ids, groups=unique_groups):
            train_groups = set(unique_groups[train_group_idx])
            val_groups = set(unique_groups[val_group_idx])
            train_idx = df['group_id'].isin(train_groups)
            val_idx = df['group_id'].isin(val_groups)
            yield np.where(train_idx)[0], np.where(val_idx)[0]
    except Exception as e:
        # fallback: GroupKFold (not stratified) - warn user
        print('WARNING: StratifiedGroupKFold not available; using GroupKFold fallback (may produce slightly imbalanced folds).')
        gkf = GroupKFold(n_splits=n_splits)
        for train_idx, val_idx in gkf.split(df, groups=groups):
            yield train_idx, val_idx
```

## Generalized Preprocessing for all models

```python
import re, string

def classical_text_preprocess(text, do_lower=True, remove_punct=False):
    t = str(text)
    if do_lower:
        t = t.lower()
    t = re.sub(r'\s+', ' ', t).strip()
    if remove_punct:
        t = t.translate(str.maketrans('', '', string.punctuation))
    return t

# Apply for classical models: create a column 'text_classical'
trainval_df['text_classical'] = trainval_df['text'].apply(lambda x: classical_text_preprocess(x, do_lower=True, remove_punct=False))
final_test_df['text_classical'] = final_test_df['text'].apply(lambda x: classical_text_preprocess(x, do_lower=True, remove_punct=False))
```

## **Specifying Metrics**
####**Saved for each run**

```python
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_recall_fscore_support, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def compute_and_save_metrics(y_true, y_pred, labels, save_prefix):
    # labels: list of label names in order of id (id2label)
    macro_f1 = f1_score(y_true, y_pred, average='macro')
    micro_f1 = f1_score(y_true, y_pred, average='micro')
    acc = accuracy_score(y_true, y_pred)
    macro_recall = recall_score(y_true, y_pred, average='macro')
    print(f'{save_prefix} -> macro_f1: {macro_f1:.4f} micro_f1: {micro_f1:.4f} acc: {acc:.4f} macro_recall: {macro_recall:.4f}')
    cr = classification_report(y_true, y_pred, target_names=labels, digits=4)
    print(cr)
    # per-class metrics
    p,r,f,_ = precision_recall_fscore_support(y_true, y_pred, labels=list(range(len(labels))))
    per = pd.DataFrame({'label':labels, 'precision':p, 'recall':r, 'f1':f})
    per.to_csv(os.path.join(WORKDIR, f'{save_prefix}_per_class.csv'), index=False)
    # confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12,10))
    sns.heatmap(cm, cmap='Blues', xticklabels=labels, yticklabels=labels, annot=False)
    plt.xlabel('Predicted'); plt.ylabel('True'); plt.title(f'Confusion matrix - {save_prefix}')
    plt.tight_layout()
    plt.savefig(os.path.join(WORKDIR, f'{save_prefix}_confusion.png'))
    plt.close()
    # return summary
    return {'macro_f1':macro_f1, 'micro_f1':micro_f1, 'accuracy':acc, 'macro_recall':macro_recall}
```

## Model:

---
#### 10 models have been used
###### Functions for each model to use CV


```python
def run_cv_for_model(trainval_df, train_fn, n_splits=5, **kwargs):
    fold_results = []
    for fold, (train_idx, val_idx) in enumerate(get_stratified_group_kfold_splits(trainval_df, n_splits=n_splits), start=1):
        tr = trainval_df.iloc[train_idx].reset_index(drop=True)
        va = trainval_df.iloc[val_idx].reset_index(drop=True)
        print(f'Running fold {fold} with train {len(tr)} val {len(va)}')
        res = train_fn(tr, va, fold=fold, **kwargs)
        res['fold'] = fold
        fold_results.append(res)
    return fold_results
```

#### Model 1
## **TF-IDF (uni+bi) + Logistic Regression**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

def train_tfidf_lr_fold(train_df, val_df, fold=0, max_features=20000, ngram_range=(1,2), C=1.0):
    out_dir = os.path.join(WORKDIR, f'tfidf_lr_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)
    # fit vectorizer on train only (use text_classical)
    vect = TfidfVectorizer(ngram_range=ngram_range, max_features=max_features)
    X_tr = vect.fit_transform(train_df['text_classical'])
    X_va = vect.transform(val_df['text_classical'])
    y_tr = train_df['label'].map(text_to_group.get) if False else train_df['label'].map({l:i for i,l in enumerate(sorted(df['label'].unique()))}).values
    # above map uses overall label2id mapping:
    label_list = sorted(df['label'].unique())
    label2id = {lab:i for i,lab in enumerate(label_list)}
    y_tr = train_df['label'].map(label2id).values
    y_va = val_df['label'].map(label2id).values
    clf = LogisticRegression(max_iter=2000, solver='saga', class_weight='balanced', C=C, random_state=SEED)
    clf.fit(X_tr, y_tr)
    preds = clf.predict(X_va)
    try:
        probs = clf.predict_proba(X_va)
    except:
        probs = None
    metrics = compute_and_save_metrics(y_va, preds, labels=label_list, save_prefix=f'tfidf_lr_fold{fold}')
    # save artifacts
    import joblib
    joblib.dump(vect, os.path.join(out_dir,'vect.joblib'))
    joblib.dump(clf, os.path.join(out_dir,'clf.joblib'))
    return {'model':'tfidf_lr','val_preds':preds,'val_probs':probs,'val_metrics':metrics,'artifact':out_dir}

# Important: TfidfVectorizer fit only on training fold. We saved vectorizer and model.
```

```python
import os
import pandas as pd
from sklearn.metrics import classification_report
import numpy as np

# --- TF-IDF + LogisticRegression (5-fold CV) ---
print("Running TF-IDF + LogisticRegression 5-fold CV (this will take a few minutes)...")
tfidf_folds = run_cv_for_model(trainval_df, train_tfidf_lr_fold, n_splits=5)

# aggregate mean/std of macro F1
tfidf_macro_f1s = [f['val_metrics']['macro_f1'] for f in tfidf_folds]
mean_f1 = np.mean(tfidf_macro_f1s); std_f1 = np.std(tfidf_macro_f1s)
print(f'TF-IDF LR mean macro-F1: {mean_f1:.4f}  std: {std_f1:.4f}')

# summary row
summary_row = {
    'run_name':'tfidf_lr_5fold',
    'model':'tfidf_lr',
    'config':"{'ngram':(1,2),'max_features':20000}",
    'val_macro_f1': mean_f1,
    'val_macro_f1_std': std_f1,
    'val_micro_f1': np.mean([f['val_metrics']['micro_f1'] for f in tfidf_folds]) if 'micro_f1' in tfidf_folds[0]['val_metrics'] else None,
    'notes':'Baseline TF-IDF + LR'
}

# --- Save to results_summary.csv ---
resfile = os.path.join(WORKDIR, 'results_summary.csv')

if os.path.exists(resfile):
    df_res = pd.read_csv(resfile)
    df_res = pd.concat([df_res, pd.DataFrame([summary_row])], ignore_index=True)
else:
    df_res = pd.DataFrame([summary_row])

df_res.to_csv(resfile, index=False)
print('Saved TF-IDF summary to', resfile)
```

#### Model 3:
## **SBERT embeddings (sentence-transformers) + Logistic Regression**

---

Produces Semantic Embeddings. We precompute Embeddings for train & val (fast), then train LR on them.

```python
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression

# load SBERT model (small & fast): 'all-MiniLM-L6-v2' or larger 'sentence-transformers/all-mpnet-base-v2'
SBERT_MODEL_NAME = 'all-MiniLM-L6-v2'  # fast; replace with 'all-mpnet-base-v2' for better quality

def train_sbert_lr_fold(train_df, val_df, fold=0, sbert_model_name=SBERT_MODEL_NAME, C=1.0):
    out_dir = os.path.join(WORKDIR, f'sbert_lr_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)
    model = SentenceTransformer(sbert_model_name)
    # compute embeddings (train only fit is not needed)
    X_tr = model.encode(train_df['text'].tolist(), show_progress_bar=True, convert_to_numpy=True)
    X_va = model.encode(val_df['text'].tolist(), show_progress_bar=True, convert_to_numpy=True)
    label_list = sorted(df['label'].unique())
    label2id = {lab:i for i,lab in enumerate(label_list)}
    y_tr = train_df['label'].map(label2id).values
    y_va = val_df['label'].map(label2id).values
    clf = LogisticRegression(max_iter=2000, solver='saga', class_weight='balanced', C=C, random_state=SEED)
    clf.fit(X_tr, y_tr)
    preds = clf.predict(X_va)
    probs = clf.predict_proba(X_va)
    metrics = compute_and_save_metrics(y_va, preds, labels=label_list, save_prefix=f'sbert_lr_fold{fold}')
    import joblib
    joblib.dump(clf, os.path.join(out_dir,'sbert_lr.joblib'))
    # optionally save embeddings for quick inference
    import numpy as np
    np.save(os.path.join(out_dir,'X_tr.npy'), X_tr)
    np.save(os.path.join(out_dir,'X_va.npy'), X_va)
    return {'model':'sbert_lr','val_preds':preds,'val_probs':probs,'val_metrics':metrics,'artifact':out_dir}
```

```python
# 2. SBERT embeddings + LogisticRegression (5-fold CV)
print("Running SBERT (sentence-transformers) + LogisticRegression CV...")
sbert_folds = run_cv_for_model(trainval_df, train_sbert_lr_fold, n_splits=5)

import numpy as np
sbert_macro_f1s = [f['val_metrics']['macro_f1'] for f in sbert_folds]
mean_f1 = np.mean(sbert_macro_f1s); std_f1 = np.std(sbert_macro_f1s)
print(f'SBERT+LR mean macro-F1: {mean_f1:.4f}  std: {std_f1:.4f}')

# Append summary
summary_row = {
    'run_name':'sbert_lr_5fold',
    'model':'sbert_lr',
    'config':f"{{'sbert_model':'{SBERT_MODEL_NAME}'}}",
    'val_macro_f1': mean_f1,
    'val_macro_f1_std': std_f1,
    'val_micro_f1': np.mean([f['val_metrics']['micro_f1'] for f in sbert_folds]) if 'micro_f1' in sbert_folds[0]['val_metrics'] else None,
    'notes':'SBERT embeddings + LR'
}
df_res = pd.read_csv(resfile)
df_res = pd.concat([df_res, pd.DataFrame([summary_row])], ignore_index=True)
df_res.to_csv(resfile, index=False)
print('Saved SBERT summary to', resfile)
```

#### Model 4:
## **Ensemble (TF-IDF+LR + SBERT+LR)**

---

Stacked meta-classifier: Using validation predictions from each base model as features and training a small Logistic Regression meta-classifier.

```python
from sklearn.linear_model import LogisticRegression

def train_ensemble_stack_fold(train_df, val_df, fold=0):
    out_dir = os.path.join(WORKDIR, f'ensemble_stack_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)
    # 1) Train TF-IDF + LR
    tf_res = train_tfidf_lr_fold(train_df, val_df, fold=fold)
    # load artifacts (vectorizer & clf)
    import joblib
    vect = joblib.load(os.path.join(tf_res['artifact'],'vect.joblib'))
    clf_tfidf = joblib.load(os.path.join(tf_res['artifact'],'clf.joblib'))
    X_va_tfidf = vect.transform(val_df['text_classical'])
    probs_tfidf = clf_tfidf.predict_proba(X_va_tfidf)
    # 2) Train SBERT + LR
    sbert_res = train_sbert_lr_fold(train_df, val_df, fold=fold)
    # load sbert clf and its X_va (we saved X_va earlier)
    X_va_sbert = np.load(os.path.join(sbert_res['artifact'],'X_va.npy'))
    clf_sbert = joblib.load(os.path.join(sbert_res['artifact'],'sbert_lr.joblib'))
    probs_sbert = clf_sbert.predict_proba(X_va_sbert)
    # 3) Stack: meta-features are concatenation of probs
    meta_X_val = np.hstack([probs_tfidf, probs_sbert])
    y_val = val_df['label'].map({lab:i for i,lab in enumerate(sorted(df['label'].unique()))}).values
    # Train meta-classifier on base models' validation predictions? That would leak. Instead:
    # Proper stacking requires out-of-fold predictions for training the meta model.
    # For simplicity here we will train a logistic meta on the train split using base models' predictions on train.
    # So compute base model probs on train as well:
    X_tr_tfidf = vect.transform(train_df['text_classical'])
    probs_tr_tfidf = clf_tfidf.predict_proba(X_tr_tfidf)
    X_tr_sbert = np.load(os.path.join(sbert_res['artifact'],'X_tr.npy'))
    probs_tr_sbert = clf_sbert.predict_proba(X_tr_sbert)
    meta_X_train = np.hstack([probs_tr_tfidf, probs_tr_sbert])
    y_tr = train_df['label'].map({lab:i for i,lab in enumerate(sorted(df['label'].unique()))}).values
    meta_clf = LogisticRegression(max_iter=2000, class_weight='balanced')
    meta_clf.fit(meta_X_train, y_tr)
    # Predict on val
    meta_preds = meta_clf.predict(meta_X_val)
    meta_probs = meta_clf.predict_proba(meta_X_val)
    metrics = compute_and_save_metrics(y_val, meta_preds, labels=sorted(df['label'].unique()), save_prefix=f'ensemble_stack_fold{fold}')
    # Save artifacts
    import joblib
    joblib.dump(meta_clf, os.path.join(out_dir,'meta_clf.joblib'))
    return {'model':'ensemble_stack','val_preds':meta_preds,'val_probs':meta_probs,'val_metrics':metrics,'artifact':out_dir}
```

```python
# 5. Ensemble stacking (TF-IDF + SBERT) 5-fold CV (this will call base train_fns again)
print("Running ensemble stacking CV (TF-IDF + SBERT) ...")
ensemble_folds = run_cv_for_model(trainval_df, train_ensemble_stack_fold, n_splits=5)

import numpy as np
ens_macro_f1s = [f['val_metrics']['macro_f1'] for f in ensemble_folds]
mean_f1 = np.mean(ens_macro_f1s); std_f1 = np.std(ens_macro_f1s)
print(f'Ensemble (stack) mean macro-F1: {mean_f1:.4f}  std: {std_f1:.4f}')

summary_row = {
    'run_name':'ensemble_stack_5fold',
    'model':'ensemble_stack',
    'config':"{'base':['tfidf_lr','sbert_lr']}",
    'val_macro_f1': mean_f1,
    'val_macro_f1_std': std_f1,
    'val_micro_f1': np.mean([f['val_metrics']['micro_f1'] for f in ensemble_folds]) if 'micro_f1' in ensemble_folds[0]['val_metrics'] else None,
    'notes':'Stacking ensemble of TF-IDF and SBERT'
}
df_res = pd.read_csv(resfile)
df_res = pd.concat([df_res, pd.DataFrame([summary_row])], ignore_index=True)
df_res.to_csv(resfile, index=False)
print('Saved ensemble summary to', resfile)
```

```python
# === TF-IDF + Multinomial Naive Bayes (per-fold train_fn) ===
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import joblib, os, numpy as np

def train_tfidf_nb_fold(train_df, val_df, fold=0, max_features=20000, ngram_range=(1,2)):
    """
    Train TF-IDF -> MultinomialNB on a single fold (train_df/val_df).
    Returns dict with val_preds, val_probs (if available), val_metrics, artifact path.
    """
    out_dir = os.path.join(WORKDIR, f'tfidf_nb_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)
    # Fit TF-IDF on train fold only (text_classical used for classical pipelines)
    vect = TfidfVectorizer(ngram_range=ngram_range, max_features=max_features)
    X_tr = vect.fit_transform(train_df['text_classical'])
    X_va = vect.transform(val_df['text_classical'])
    # consistent label mapping (use sorted labels from original df)
    label_list = sorted(df['label'].unique())
    label2id = {lab:i for i,lab in enumerate(label_list)}
    y_tr = train_df['label'].map(label2id).values
    y_va = val_df['label'].map(label2id).values
    clf = MultinomialNB()
    clf.fit(X_tr, y_tr)
    preds = clf.predict(X_va)
    try:
        probs = clf.predict_proba(X_va)
    except:
        probs = None
    metrics = compute_and_save_metrics(y_va, preds, labels=label_list, save_prefix=f'tfidf_nb_fold{fold}')
    # Save artifacts
    joblib.dump(vect, os.path.join(out_dir,'vect.joblib'))
    joblib.dump(clf, os.path.join(out_dir,'clf_nb.joblib'))
    return {'model':'tfidf_nb','val_preds':preds,'val_probs':probs,'val_metrics':metrics,'artifact':out_dir}
```

#### Model 6:
## **Keras Dense model (TextVectorization + Embedding + Dense head)**

---

A simple neural model using tokenization and an embedding layer.

```python
# === Keras Dense model per-fold train_fn ===
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization, Embedding, GlobalAveragePooling1D, Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np, joblib, os, json

def train_keras_dense_fold(train_df, val_df, fold=0, max_vocab=10000, max_len=45, emb_dim=128, epochs=20, batch_size=32):
    out_dir = os.path.join(WORKDIR, f'keras_dense_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)
    # labels
    label_list = sorted(df['label'].unique())
    label2id = {lab:i for i,lab in enumerate(label_list)}
    y_tr = train_df['label'].map(label2id).values
    y_va = val_df['label'].map(label2id).values
    # One-hot for Keras fit
    y_tr_ohe = tf.keras.utils.to_categorical(y_tr, num_classes=len(label_list))
    y_va_ohe = tf.keras.utils.to_categorical(y_va, num_classes=len(label_list))
    # TextVectorization per-fold (fit/adapt on train only)
    text_vectorizer = TextVectorization(max_tokens=max_vocab, output_mode='int', output_sequence_length=max_len)
    text_vectorizer.adapt(train_df['text'].astype(str).values)
    vocab_size = int(text_vectorizer.vocabulary_size())
    # Build model: string input -> vectorize -> embedding -> pooling -> dense softmax
    inp = Input(shape=(1,), dtype=tf.string, name='text_input')
    x = text_vectorizer(inp)
    x = Embedding(input_dim=vocab_size, output_dim=emb_dim, input_length=max_len, name='embedding')(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation='relu')(x)
    out = Dense(len(label_list), activation='softmax')(x)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    # callbacks
    es = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=1)
    # Train
    history = model.fit(train_df['text'].astype(str).values, y_tr_ohe,
                        validation_data=(val_df['text'].astype(str).values, y_va_ohe),
                        epochs=epochs, batch_size=batch_size, callbacks=[es], verbose=2)
    # Predict on val
    probs = model.predict(val_df['text'].astype(str).values, batch_size=64)
    preds = np.argmax(probs, axis=1)
    # compute metrics and save
    metrics = compute_and_save_metrics(y_va, preds, labels=label_list, save_prefix=f'keras_dense_fold{fold}')
    # save model and vectorizer
    model.save(os.path.join(out_dir, 'keras_dense_model.keras'))
    # Can't directly save TF TextVectorization with joblib — save its vocab
    vocab = text_vectorizer.get_vocabulary()
    with open(os.path.join(out_dir,'text_vectorizer_vocab.json'),'w') as f:
        json.dump(vocab, f)
    return {'model':'keras_dense','val_preds':preds,'val_probs':probs,'val_metrics':metrics,'artifact':out_dir}
```

#### Model 7:
## **Universal Sentence Encoder (USE) transfer-learning model**

---

```python
# === USE (Universal Sentence Encoder) per-fold train_fn ===
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np, os, json

# Cache the USE model object globally to avoid reloading each fold
USE_MODEL_HANDLE = "https://tfhub.dev/google/universal-sentence-encoder/4"
try:
    base_use_layer = hub.load(USE_MODEL_HANDLE)  # load TF Hub module directly
    print("USE loaded.")
except Exception as e:
    print("Error loading USE:", e)
    raise

# Wrap into a Keras-compatible layer
class USELayer(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.use = base_use_layer

    def call(self, inputs):
        return self.use(inputs)

def train_use_fold(train_df, val_df, fold=0, dense_units=[512,128], epochs=10, batch_size=32):
    out_dir = os.path.join(WORKDIR, f'use_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)

    # safer: build label list from available data
    label_list = sorted(pd.concat([train_df['label'], val_df['label']]).unique())
    label2id = {lab:i for i,lab in enumerate(label_list)}
    y_tr = train_df['label'].map(label2id).values
    y_va = val_df['label'].map(label2id).values
    y_tr_ohe = tf.keras.utils.to_categorical(y_tr, num_classes=len(label_list))
    y_va_ohe = tf.keras.utils.to_categorical(y_va, num_classes=len(label_list))

    # Build model: USE (frozen) -> Dense -> Dense -> softmax
    inp = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text_input')
    x = USELayer()(inp)   # ✅ now wrapped layer is Keras-compatible
    for units in dense_units:
        x = tf.keras.layers.Dense(units, activation='relu')(x)
        x = tf.keras.layers.Dropout(0.2)(x)
    out = tf.keras.layers.Dense(len(label_list), activation='softmax')(x)

    model = tf.keras.Model(inputs=inp, outputs=out)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    es = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=3, restore_best_weights=True, verbose=1
    )

    history = model.fit(
        train_df['text'].astype(str).values, y_tr_ohe,
        validation_data=(val_df['text'].astype(str).values, y_va_ohe),
        epochs=epochs, batch_size=batch_size,
        callbacks=[es], verbose=2
    )

    probs = model.predict(val_df['text'].astype(str).values, batch_size=64)
    preds = np.argmax(probs, axis=1)

    metrics = compute_and_save_metrics(
        y_va, preds, labels=label_list,
        save_prefix=f'use_fold{fold}'
    )

    # save model
    model.save(os.path.join(out_dir, 'use_model.keras'))

    return {
        'model':'use',
        'val_preds':preds,
        'val_probs':probs,
        'val_metrics':metrics,
        'artifact':out_dir
    }
```

```python
# RUN USE CV (5-fold)
print("Running USE (Universal Sentence Encoder) 5-fold CV ...")
use_folds = run_cv_for_model(trainval_df, train_use_fold, n_splits=5)

import numpy as np
use_macro_f1s = [f['val_metrics']['macro_f1'] for f in use_folds]
mean_f1 = np.mean(use_macro_f1s); std_f1 = np.std(use_macro_f1s)
print(f'USE mean macro-F1: {mean_f1:.4f}  std: {std_f1:.4f}')

# Append to results_summary.csv
resfile = os.path.join(WORKDIR, 'results_summary.csv')
summary_row = {
    'run_name':'use_5fold',
    'model':'use',
    'config':"{'dense_units':[512,128],'trainable_encoder':False}",
    'val_macro_f1': mean_f1,
    'val_macro_f1_std': std_f1,
    'val_micro_f1': np.mean([f['val_metrics'].get('micro_f1', np.nan) for f in use_folds]),
    'notes':'Universal Sentence Encoder (frozen) + classifier head'
}
df_res = pd.read_csv(resfile)
df_res = pd.concat([df_res, pd.DataFrame([summary_row])], ignore_index=True)
df_res.to_csv(resfile, index=False)
print('Saved USE summary to', resfile)
```

## **CNN 1D with pretrained GloVe embeddings (Keras)**

```python
# Download GloVe 100d (if not already)
GLOVE_DIR = os.path.join(WORKDIR, 'glove')
os.makedirs(GLOVE_DIR, exist_ok=True)
glove_path = os.path.join(GLOVE_DIR, 'glove.6B.100d.txt')
if not os.path.exists(glove_path):
    print('Downloading GloVe 100d (approx 800MB zip; we extract only 100d if needed)...')
    !wget -q -nc http://nlp.stanford.edu/data/glove.6B.zip -P {GLOVE_DIR}
    !unzip -n -q {os.path.join(GLOVE_DIR,"glove.6B.zip")} -d {GLOVE_DIR}


import os
import numpy as np
import pandas as pd
# Keras model
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPool1D, Dense, Dropout, Input
from tensorflow.keras.models import Model
import joblib

def build_embedding_matrix(tokenizer, glove_path, emb_dim=100):
    embeddings_index = {}
    with open(glove_path, 'r', encoding='utf8') as f:
        for line in f:
            values = line.strip().split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    vocab_size = min(len(tokenizer.word_index) + 1, tokenizer.num_words)
    embedding_matrix = np.zeros((vocab_size, emb_dim))
    for word, i in tokenizer.word_index.items():
        if i >= tokenizer.num_words:
            continue
        vec = embeddings_index.get(word)
        if vec is not None:
            embedding_matrix[i] = vec
    return embedding_matrix

def train_cnn_glove_fold(train_df, val_df, fold=0, max_words=20000, max_len=128, emb_dim=100, epochs=6, batch=32):
    out_dir = os.path.join(WORKDIR, f'cnn_glove_fold{fold}')
    os.makedirs(out_dir, exist_ok=True)
    # tokenizer fit on train only
    tok = Tokenizer(num_words=max_words, oov_token='<OOV>')
    tok.fit_on_texts(train_df['text_classical'])
    def prep(df):
        seq = tok.texts_to_sequences(df['text_classical'])
        return pad_sequences(seq, maxlen=max_len)
    X_tr = prep(train_df); X_va = prep(val_df)
    label_list = sorted(df['label'].unique())
    label2id = {lab:i for i,lab in enumerate(label_list)}
    y_tr = tf.keras.utils.to_categorical(train_df['label'].map(label2id).values, num_classes=len(label_list))
    y_va = tf.keras.utils.to_categorical(val_df['label'].map(label2id).values, num_classes=len(label_list))
    # embedding matrix
    tokenizer_wrapper = tok
    tokenizer_wrapper.num_words = max_words
    emb_matrix = build_embedding_matrix(tok, glove_path, emb_dim=emb_dim)
    # model
    inp = Input(shape=(max_len,))
    x = Embedding(input_dim=emb_matrix.shape[0], output_dim=emb_dim, weights=[emb_matrix], trainable=False)(inp)
    x = Conv1D(128, 5, activation='relu')(x)
    x = GlobalMaxPool1D()(x)
    x = Dropout(0.2)(x)
    out = Dense(len(label_list), activation='softmax')(x)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(X_tr, y_tr, validation_data=(X_va, y_va), epochs=epochs, batch_size=batch, verbose=2)
    preds = model.predict(X_va).argmax(axis=1)
    metrics = compute_and_save_metrics(val_df['label'].map(label2id).values, preds, labels=label_list, save_prefix=f'cnn_glove_fold{fold}')
    model.save(os.path.join(out_dir,'model.h5'))
    joblib.dump(tok, os.path.join(out_dir,'tokenizer.joblib'))
    return {'model':'cnn_glove','val_preds':preds,'val_probs':None,'val_metrics':metrics,'artifact':out_dir}
```
