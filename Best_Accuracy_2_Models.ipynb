## Common Preprocessing

```python
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import LabelEncoder

# Map labels to integers
le = LabelEncoder()
df['label_id'] = le.fit_transform(df['label'])
id2label = {i:l for i,l in enumerate(le.classes_)}
num_labels = len(le.classes_)
print('Number of classes:', num_labels)

# Simple train-test split (stratified)
train_df, test_df = train_test_split(df, test_size=0.15, random_state=42, stratify=df['label_id'])
print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")
```

## Pipeline 1: Transformer + XGBoost tail
---

#### **Overview:**

Encoding each symptom text with ClinicalBERT/SciBERT → pooled embeddings

Feeding embeddings to XGBoost classifier for final prediction

```python
import torch
from transformers import AutoTokenizer, AutoModel
import xgboost as xgb
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score

# Load Transformer model (pretrained)
MODEL_NAME = "emilyalsentzer/Bio_ClinicalBERT"  # ClinicalBERT
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModel.from_pretrained(MODEL_NAME)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

# Function to get sentence embeddings
def get_transformer_embeddings(texts, batch_size=16):
    all_embs = []
    with torch.no_grad():
        for i in range(0, len(texts), batch_size):
            batch = texts[i:i+batch_size]
            enc = tokenizer(batch, padding=True, truncation=True, max_length=128, return_tensors='pt').to(device)
            out = model(**enc)
            # Use [CLS] token embedding
            cls_emb = out.last_hidden_state[:,0,:].cpu().numpy()
            all_embs.append(cls_emb)
    return np.vstack(all_embs)

# Get embeddings
X_train_emb = get_transformer_embeddings(train_df['text'].tolist())
X_test_emb  = get_transformer_embeddings(test_df['text'].tolist())
y_train = train_df['label_id'].values
y_test  = test_df['label_id'].values

# Train XGBoost
xgb_clf = xgb.XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    objective='multi:softprob',
    num_class=num_labels,
    eval_metric='mlogloss',
    verbosity=1,
    n_jobs=-1
)
xgb_clf.fit(X_train_emb, y_train)

# Predict & metrics
y_pred = xgb_clf.predict(X_test_emb)
y_prob = xgb_clf.predict_proba(X_test_emb)

acc = accuracy_score(y_test, y_pred)
macro_f1 = f1_score(y_test, y_pred, average='macro')
micro_f1 = f1_score(y_test, y_pred, average='micro')
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print(f"Transformer+XGBoost Test -> acc: {acc:.4f}, macro_f1: {macro_f1:.4f}, micro_f1: {micro_f1:.4f}")

# Save results
np.savez('/content/transformer_xgb_results.npz',
         y_true=y_test, y_pred=y_pred, y_prob=y_prob)

# Copy to Drive
!cp /content/transformer_xgb_results.npz "{WORKDIR}/"
```

## Pipeline 2: TF-IDF + SBERT stacking
---

#### **Overview:**

TF-IDF + Logistic Regression channel

SBERT embeddings + Logistic Regression channel

Stack outputs with meta-classifier

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sentence_transformers import SentenceTransformer
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import numpy as np
import os

# TF-IDF
tfidf = TfidfVectorizer(ngram_range=(1,2), max_features=3000)
X_train_tfidf = tfidf.fit_transform(train_df['text'])
X_test_tfidf  = tfidf.transform(test_df['text'])

# SBERT embeddings
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')  # light & fast
X_train_sbert = sbert_model.encode(train_df['text'].tolist(), convert_to_numpy=True, batch_size=16, show_progress_bar=True)
X_test_sbert  = sbert_model.encode(test_df['text'].tolist(), convert_to_numpy=True, batch_size=16)

# Base classifiers
clf_tfidf  = LogisticRegression(max_iter=1000, multi_class='multinomial')
clf_sbert  = LogisticRegression(max_iter=1000, multi_class='multinomial')

# Stacking
stack = StackingClassifier(
    estimators=[('tfidf', clf_tfidf), ('sbert', clf_sbert)],
    final_estimator=LogisticRegression(max_iter=1000, multi_class='multinomial'),
    cv=5,
    n_jobs=-1
)

# Combine features
X_train_combined = np.hstack([X_train_tfidf.toarray(), X_train_sbert])
X_test_combined  = np.hstack([X_test_tfidf.toarray(), X_test_sbert])

stack.fit(X_train_combined, y_train)

# Predict & metrics
y_pred_stack = stack.predict(X_test_combined)
y_prob_stack = stack.predict_proba(X_test_combined)

acc_stack = accuracy_score(y_test, y_pred_stack)
macro_f1_stack = f1_score(y_test, y_pred_stack, average='macro')
micro_f1_stack = f1_score(y_test, y_pred_stack, average='micro')
precision_stack = precision_score(y_test, y_pred_stack, average='macro')
recall_stack = recall_score(y_test, y_pred_stack, average='macro')

print(f"TF-IDF + SBERT Stacking Test -> acc: {acc_stack:.4f}, macro_f1: {macro_f1_stack:.4f}, micro_f1: {micro_f1_stack:.4f}, precision: {precision_stack:.4f}, recall: {recall_stack:.4f}")

# Save results locally first
np.savez('/content/tfidf_sbert_stack_results.npz',
         y_true=y_test, y_pred=y_pred_stack, y_prob=y_prob_stack)

# Copy to Drive
!cp /content/tfidf_sbert_stack_results.npz "{WORKDIR}/"
```
